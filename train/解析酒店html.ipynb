```python
from bs4 import BeautifulSoup

# 打开并读取 HTML 文件
with open("秦皇岛.com.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(html_content, "html.parser")

# 选择目标元素
attraction_detail = soup.select('div.hotel-list-page')

# 格式化并打印 HTML 内容
if attraction_detail:
    for detail in attraction_detail:
        print(detail.prettify())  # 使用 prettify() 方法格式化输出
else:
    print("未找到目标元素")
```
